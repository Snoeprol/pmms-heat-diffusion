```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filepath = 'output4_1000.txt'

data1000 = {}
data3000 = {}
data5000 = {}
data7000 = {}
datas = [data1000, data3000, data5000, data7000]

for data in datas:
    # fill data with arrays
    with open (filepath) as f:
        line = f.readline().strip()
        while line:
            if (line[0] != "C" and not "e" in line):
                if not (float(line) in data):
                    data[float(line)] = []


            line = f.readline().strip()




filepath = 'output4_1000.txt'
with open(filepath) as f:
   line = f.readline().strip()
   while line:
        if not "e+" in line and not "e-" in line and line[0] != "C":
            bufsize = int(line)
        
        elif line[0] == "C":
           if int(line[-1]) != 1:
               print(f"Found incorrect output: {line}")
               break;
        else:
            time = float(line)
            data1000[bufsize].append(time)


        line = f.readline().strip()

filepath = 'output5_3000.txt'
with open(filepath) as f:
   line = f.readline().strip()
   while line:
        if not "e+" in line and not "e-" in line and line[0] != "C":
            bufsize = int(line)
        
        elif line[0] == "C":
           if int(line[-1]) != 1:
               print(f"Found incorrect output: {line}")
               break;
        else:
            time = float(line)
            data3000[bufsize].append(time)


        line = f.readline().strip()

filepath = 'output3_5000.txt'
with open(filepath) as f:
   line = f.readline().strip()
   while line:
        if not "e+" in line and not "e-" in line and line[0] != "C":
            bufsize = int(line)
        
        elif line[0] == "C":
           if int(line[-1]) != 1:
               print(f"Found incorrect output: {line}")
               break;
        else:
            time = float(line)
            data5000[bufsize].append(time)


        line = f.readline().strip()

filepath = 'output6_7000.txt'
with open(filepath) as f:
   line = f.readline().strip()
   while line:
        if not "e+" in line and not "e-" in line and line[0] != "C":
            bufsize = int(line)
        
        elif line[0] == "C":
           if int(line[-1]) != 1:
               print(f"Found incorrect output: {line}")
               break;
        else:
            time = float(line)
            data7000[bufsize].append(time)



        line = f.readline().strip()


```

```python
buf_sizes = data.keys()
print("bufsizes:")
print(buf_sizes)

results1000 = []
results3000 = []
results5000 = []
results7000 = []

for size in buf_sizes:
    current = data1000[size]
    avg = np.mean(current)
    results1000.append(avg)

for size in buf_sizes:
    current = data3000[size]
    avg = np.mean(current)
    results3000.append(avg)

for size in buf_sizes:
    current = data5000[size]
    avg = np.mean(current)
    results5000.append(avg)

for size in buf_sizes:
    current = data7000[size]
    avg = np.mean(current)
    results7000.append(avg)

print(results1000)
print(results5000)
print(results7000)

plt.plot(list(buf_sizes), results1000, label="Length 1000")
plt.plot(list(buf_sizes), results3000, label="Length 3000")
plt.plot(list(buf_sizes), results5000, label="Length 5000")
plt.plot(list(buf_sizes), results7000, label="Length 7000")
plt.grid()
plt.xlabel("Buffer size")
plt.ylabel("Execution time [s]")
plt.title("Pipie line sort execution times for different lengths\nand different buffer sizes")
plt.legend()
plt.show()

```

```python
filepath = 'output2.txt'
data = {}
# fill data with arrays
with open (filepath) as f:
    line = f.readline().strip()
    while line:
        if (line[0] != "C" and not "e" in line):
            if not (float(line) in data):
                data[float(line)] = []


        line = f.readline().strip()
        # print(line)
print(data)

with open(filepath) as f:
   line = f.readline().strip()
   while line:
        if not "e+" in line and not "e-" in line and line[0] != "C":
            bufsize = int(line)
        
        elif line[0] == "C":
           if int(line[-1]) != 1:
               print(f"Found incorrect output: {line}")
               break;
        else:
            time = float(line)
            data[bufsize].append(time)


        line = f.readline().strip()

print(data)
```

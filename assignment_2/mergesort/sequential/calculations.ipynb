```python
file_seq = open ('output.txt')
lines_seq = file_seq.readlines()
lengths = ['10000', '100000', '1000000', '10000000', '100000000', '1000000000']
ascending = [[] for i in range(6)]
descending = [[] for i in range(6)]
random = [[] for i in range(6)]
for i in range(0, len(lines_seq), 3):
    # print(lines_seq[i+1])
    if lines_seq[i+1].strip() == 'a':
        ascending[lengths.index(lines_seq[i].strip())].append(float(lines_seq[i+2].strip()))
    elif lines_seq[i+1].strip() == 'r':
        descending[lengths.index(lines_seq[i].strip())].append(float(lines_seq[i+2].strip()))
    elif lines_seq[i+1].strip() == 'd':
        random[lengths.index(lines_seq[i].strip())].append(float(lines_seq[i+2].strip()))
    
```

```python
# print(ascending)
# print(descending)
# print(random)
def avg(lst): 
    return sum(lst) / len(lst) 
```

```python
import matplotlib.pyplot as plt
import numpy as np
asc_avg = [avg(x) for x in ascending]
desc_avg = [avg(x) for x in descending]
rand_avg = [avg(x) for x in random]
int_lengths = [int(x) for x in lengths]
print(int_lengths)

plt.plot(int_lengths, asc_avg)
plt.plot(int_lengths, desc_avg)
plt.plot(int_lengths, rand_avg)
plt.xlabel("Length of vector [-]")
plt.ylabel("Execution time [s]")
plt.show()
```

```python
print("16 threads:")

file_seq = open ('../parallel/output.txt')
lines_seq = file_seq.readlines()
lengths = ['10000', '100000', '1000000', '10000000', '100000000', '1000000000']
ascending = [[] for i in range(6)]
descending = [[] for i in range(6)]
random = [[] for i in range(6)]
for i in range(0, len(lines_seq), 3):
    if lines_seq[i+1].strip() == "16":
        # print(lines_seq[i+1])
        if lines_seq[i+2].strip() == 'a':
            ascending[lengths.index(lines_seq[i].strip())].append(float(lines_seq[i+3].strip()))
        elif lines_seq[i+2].strip() == 'r':
            descending[lengths.index(lines_seq[i].strip())].append(float(lines_seq[i+3].strip()))
        elif lines_seq[i+2].strip() == 'd':
            random[lengths.index(lines_seq[i].strip())].append(float(lines_seq[i+3].strip()))

import matplotlib.pyplot as plt
asc_avg = [avg(x) for x in ascending]
desc_avg = [avg(x) for x in descending]
rand_avg = [avg(x) for x in random]
int_lengths = [int(x) for x in lengths]
print(int_lengths)
print(asc_avg)

plt.plot(int_lengths, asc_avg, label="Ascending")
plt.plot(int_lengths, desc_avg, label="Descending")
plt.plot(int_lengths, rand_avg, label="Random")
plt.scatter(int_lengths, asc_avg, label="Ascending")
plt.scatter(int_lengths, desc_avg, label="Descending")
plt.scatter(int_lengths, rand_avg, label="Random")
plt.xscale("log")
plt.yscale("log")
plt.legend()
plt.xlabel("Length of vector [-]")
plt.ylabel("Execution time [s]")
plt.tight_layout()
plt.savefig("mergebench.png")
plt.show()
```
